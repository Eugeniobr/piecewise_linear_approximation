# Definição do Problema

Queremos resolver o seguinte problema:
\begin{align*}
& \min x^2 + 12x + 36 \\
& \quad \quad x \in \mathbb{R}
\end{align*}


# Modelagem

O problema colocado tem solução analítica, que pode ser facilmente obtida. No entanto, aqui vamos por outro caminho. A maneira que escolhelhos aqui para resolver esse problema é aproximar a função $x^2$ por uma série de funções lineares e que definam uma região convexa. Feita essa aproximação, constrói-se o modelo de otimização com as funções aproximadoras. Supondo-se que as funções aproximadoras sejam 
\begin{align*}
& c_1x + d_1,\\
& c_2x + d_2, \\
& \quad \vdots  \\
& c_Nx + d_N
\end{align*},

em que $N$ é um natural e $c_i, d_i$ são, respectivamente, os coenficientes angular e linear das $N$ funções, constrói-se o seguinte modelo de otimização com as funções aproximadoras:

\begin{align}
&\min \max_{i=1,...,N}(c_ix + d_i)\\
& \quad x \in \mathbb{R}
\end{align}

O modelo acima, apesar de correto, apresenta um empicilho para que sua otimização seja realizada: a função $max$. Para retirar esse empecilho, reescrevemos o modelo e adicionamos uma nova variável, que designamos aqui por $z$. O modelo reescrito é

\begin{align}
&\min z \\
& z \geq c_1x + d_1 \\
& z \geq c_2x + d_2 \\
& \quad \vdots  \\
& z \geq c_Nx + d_N \\
& \quad x \in \mathbb{R}
\end{align}


De posse desse modelo, resta agora sabermos quais funções podemos utilizar para aproximar o polinômio $f(x)$. Uma alternativa seria usarmos retas que tangenciam o polinômio $f(x)$, com o objetivo de construirmos uma envoltoria convexa para essa função. Pensando em retas que tangenciam, lembramos de derivadas. Portanto, uma maneira fácil de pensarmos em funções aproximadoras é determinar as retas cujos coeficientes angulares sejam derivadas de $f(x)$ e que tangenciem essa função. Vejamos como determinar essas retas. Seja
$$ \Bigl( \frac{df(x_1)}{dx}, \frac{df(x_2)}{dx}, \cdots, \frac{df(x_N)}{dx} \Bigr)$$

o vetor cujo o $i$-ésimo compontente $\frac{df(x_i)}{dx}$ é a derivada no ponto $x_i$. Supondo que $b_i$ seja o coeficente linear da reta cujo coeficiente angular é $\frac{df(x_i)}{dx}$ e tangencia $f(x)$ em $x_i$, determinamos $b_i$ como 
$$ b_i = f(x_i) - \frac{df(x_i)}{dx}x_i $$

Sabendo que a derivada de $f(x)$ é 
$$ \frac{df(x)}{x} = 2x + 12 $$
e levando em conta toda a modelagem apresentada até agora, estamos prontos para resolver o problema.

# Construção do Polinômio $f(x)$ e das funções aproximadoras

Sabe-se que a função $f(x) = x^2 + 12x + 36$ que está definida em todo $\mathbb{R}$. Porém, aqui restringimos o nosso problema ao intervalo $[-8, -4]$. Como primeiros passos, criamos os valores discretizamos o intervalo considerado e calculamos respectivos valores de $f(x)$.

```python
x = np.arange(-8, -3.99, 0.01)
```

```python
y = x**2 + 12*x + 36
```

Para se ter uma ideia visual da função, ''plotamos'' o gráfico de $f(x)$ no intervalo.

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico do polinômio')
```

Agora, calculamos os valores da derivada para todos os pontos do intervalo discretizado.

```python
derivative = 2*x + 12
```

Em seguida, calculamos os coeficientes lineares das retas que tangenciam nossa função.

```python
b = y - derivative*x 
```

Como forma de dar ideia da construção da região convexa que queremos desenvolver para $f(x)$, criamos as retas L1 e L2. 

```python
L1 = derivative[0]*x + b[0]
L2 = derivative[-1]*x + b[-1]
```

Para visualizar a região determinada pelas retas L1 e L2, criamos uma função que "plota" essas retas juntamente com o gráfico de $f(x)$.

```python
def plot_lines(x, y, line_list):
    
    plt.plot(x,y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Aproximação por funções lineares por partes')
    for line in line_list:
        plt.plot(x, line)
```

```python
plot_lines(x, y, [L1,L2])
```

Pelo gráfico, podemos ver que as duas retas tangenciam $f(x)$ no intervalo considerado. Para o propósito de resolver o problema de minização em questão, a região que nos interessa é justamente a região em que $y \geq L1$ e $ y \geq L2$. Nessa região, vemos que o mínimo é atingido justamente na interseção entre $L1$ e $L2$.

Ao compararmos o mínimo da parábola, que é o gráfico de $f(x)$, e o mínimo da região determinda por $L1$ e $L2$, vemos que esses mínimos parecem distantes. Para remediar esse problema, construimos mais duas retas $L3$ e $L4$, com o objetivo criar uma região convexa mais próxima do gráfico de $f(x)$. 

```python
L3 = derivative[99]*x + b[99]
L4 = derivative[-100]*x + b[-100]
```

Então, plotamos o gráfico das retas $L1,L2,L3 e L4$ junto ao gráfico de $f(x)$.

```python
plot_lines(x, y, [L1,L2,L3,L4])
```

Dessa vez, o mínimo da região convexa é obtido por
\begin{align}
& y > L1, \\
& y > L2, \\
& y > L3, \\
& y > L4.
\end{align}
Por inspeção visual, vemos que esse mínimo na interseção das retas $L3$ e $L4$. Também vemos que esse mínimo é mais próximo do mínimo de $f(x)$ quando comparado ao mínimo obtido pela intersecção das retas $L1$ e $L2$.

Há, porém, uma dificuldade que surge nessa metodologia de aproximação: nem todo conjunto de retas escolhidas para aproximar $f(x)$ serve para nosso propósito de otimização. Isso se deve ao fato de que, para determinados conjuntos de retas, o problema de otimização não possui mínimo finito. Vejamos um exemplo:

```python
L1 = derivative[0]*x + b[0]
L2 = derivative[50]*x + b[50]
```

```python
plot_lines(x, y, [L1,L2])
```

No caso das retas $L1$ e $L2$ que escolhemos agora, o problema é ilimitado inferiormente. 

# Investigando o método de aproximação

Agora, vamos para outra parte de nossa investigação. A pergunta que queremos responder é:
    
###  *À medida que aumentamos o número de retas aproximadoras, como o ótimo do problema investigado se comporta ?*
    

Para responder a essa pergunta, faremos o seguinte: Para cada $M$, o número de retas que deverão aproximar $f(x)$, resolvemos o problema de otimização que estamos estudando.

Dado um valor de $M$, o próximo passo é definir como escolher as retas que aproximarão a função $f(x)$. Aqui, será criado um procedimento em que $M$ retas distintas são escolhidas de forma aleatória, a partir dos $N$ valores de $x$ do intervalo discretizado. 

Como a escolha das $M$ é feita de forma aleatória, devemos então resolver o problema de otimização para vários conjuntos de $M$ retas, com o objetivo de obtermos valores médios do mínimo obtido. 


A seguir, estão duas funções criadas para construir instâncias do problema de otimização. O formato da instância é AMPL.

```python
def write_ampl(J, parms):
    with open('parm.dat', 'w') as f:
        
        f.write('set J:= ')
        
        for i in range(1,J+1):
            f.write(str(i) + ' ')
        f.write(';\n')
        
        
        f.write('param c:= ')
        for i in range(1,J+1):
            f.write(str(i) + ' ' + str(parms[i-1][0]) + '\n')
        f.write(';\n')
        
        f.write('param d:= ')
        for i in range(1,J+1):
            f.write(str(i) + ' ' + str(parms[i-1][1]) + '\n')
        f.write(';\n')
```

```python
def create_instance(J, derivative, b):
    
    indexes = np.random.default_rng().choice(len(derivative), J, replace=False)
        
    parms = []
    for index in indexes:
        parms.append([derivative[index], b[index]])
        
    parms = np.array(parms)
    write_ampl(parms.shape[0], parms)
    
    return parms
```

A seguir, obtemos a média do mínimo obtido para $M$ retas, em que $M \in \{2,...,15\}$. Com o objetivo de obter valores médios, resolvemos o problema de otimização para 30 conjuntos de $M$, excluindo os resultados dos conjuntos cujo problema de otimização é ilimitado inferiormente.
